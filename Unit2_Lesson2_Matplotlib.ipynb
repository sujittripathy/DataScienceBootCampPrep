```python
np.random.seed(1221)

df = pd.DataFrame()

df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2

df['counts_sq'] = df.index ** 2
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
#plotting on DataFrame
df.plot(kind='scatter',x='rand',y='rand_sq',title='Squared Chart')
```

```python
df
```

```python
plt.plot(df['rand'])
plt.show()
```

```python
plt.plot(df['rand'],color='red')
plt.plot(df['rand_shift'],color='green')
plt.ylim(-0.1,3.1)
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

```python
plt.scatter(df['rand'],
            df['rand_sq'],
           color='purple',
           marker='x', s=10)
plt.scatter(df['rand'],
            df['rand_shift'],
           color='green',
           marker='x', s=10)
plt.show()
```

# Pandas Plot

```python
df.plot(kind='scatter', x='counts_sq',y='counts_sqrt')
df.plot(kind='line')
plt.show()
```

```python
plt.figure(figsize=(20,5))

plt.subplot(2,1,1) # (row,coloumn,placement of the chart)
plt.plot(df['rand'],color='Purple')
plt.ylabel('Values')
plt.title('Random Values')

plt.subplot(2,1,2)
plt.plot(df['rand_shift'], color='green')
plt.ylabel('Shifted Values')
plt.title('Shifted Series')

plt.show()
```

```python
plt.figure(figsize=(20,5))

plt.subplot(1,2,1)
plt.plot(df['rand'],color='Purple')
plt.ylabel('Values')
plt.title('Random Values')

plt.subplot(1,2,2)
plt.scatter(x=df['rand_sq'],y=df['rand'],color='green')
plt.ylabel('Squared values')
plt.title('Squared Series')

plt.tight_layout()
plt.show()
```

# Statistical Plots

```python
x = np.random.normal(10,5,1000)

plt.hist(x)
plt.show()
```

```python
x = np.random.normal(10,5,1000)

plt.hist(x, bins=40, color='red')
plt.xlabel('Random Values')
plt.title('Histogram Random')
plt.show()
```

```python
x = np.random.normal(10,5,1000)
y = np.random.normal(10,5,10000)

plt.hist(x, color='blue', bins=np.arange(-10,40),alpha=0.5)
plt.hist(y, color='red', bins=np.arange(-10,40),alpha=0.5)
plt.title('Manually setting bin')
plt.xlabel('Random Values')

plt.show()
```

```python
x = np.random.normal(10,5,1000)
y = np.random.normal(10,5,10000)

plt.hist(x, normed=True,color='blue', bins=np.arange(-10,40),alpha=0.5)
plt.hist(y, normed=True, color='red', bins=np.arange(-10,40),alpha=0.5)
plt.title('Normed Histogran')
plt.xlabel('Random Values')

plt.show()
```
